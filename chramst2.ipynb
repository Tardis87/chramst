```python
from collections import Counter




class Chramst:
    def __init__(self):
        
        self.hraci = {
            "üîµ Modry": {"poradie": 1, "zivoty": 4, "karty na ruke": [1, 2, 3, 4, 5, 6], "odhadzovaci balicek": []},
            "üü° Zlty": {"poradie": 2, "zivoty": 4, "karty na ruke": [1, 2, 3, 4, 5, 6], "odhadzovaci balicek": []},
            "üü¢ Zeleny": {"poradie": 3, "zivoty": 4, "karty na ruke": [1, 2, 3, 4, 5, 6], "odhadzovaci balicek": []},
            "üî¥ Cerveny": {"poradie": 4, "zivoty": 4, "karty na ruke": [1, 2, 3, 4, 5, 6], "odhadzovaci balicek": []}
        }
        
        self.karty_zahrane_v_kole = []
        self.hra_bezi = True
        
    def zobraz_hracov(self):
        for meno, info in self.hraci.items():
            print(f"{meno}: {info['poradie']}. miesto, Zivoty: {info['zivoty']}. karty na ruke: {info['karty na ruke']}. karty v odhadzovacom balicku: {info['odhadzovaci balicek']} ")
         
     
        
    def zahraj_kartu(self, meno_hraca):
        # Zisti karty tohto hraca
        karty_hraca = self.hraci[meno_hraca]["karty na ruke"]
        
        while True: #ked som to tu nemala, ak zadal hrac zlu kartu, tak isiel iny hrac nasledne na to
            vstup = input(f"{meno_hraca}, vyber si kartu co chces zahrat {karty_hraca}: ak chces ukoncit hru napis: koniec ")
       
            if vstup.lower() == "koniec":
                print("Ukoncujem hru...")
                self.hra_bezi = False  # Nastav flag na False
                return  # vrati sa z funkcie a zastavi hru
                
            #ak nie je osetrene ak da hrac len enter, tak to cele padne
            if vstup == "":
                print("Nezadal si ziadnu kartu. Skus znova.")
                continue  # Vr√°≈• sa na zaciatok while loopu
        
             # Skontroluj, ci su to len cisla
            if not vstup.isdigit():
                print("To nie je cislo. Skus znova.")
                continue
            
            # ked uz viem ze to je cislo, mozme previest
            vyber_kartu = int(vstup)
        
            if vyber_kartu not in karty_hraca:
                print(f"Cislo {vyber_kartu} nie je spravne. Vyber si jednu z kariet co mas na ruke")
            else:
                #print(f"{meno_hraca} zahral kartu {vyber_kartu}") #tiez to tu uz nepotrebujem, 
                # Odstr√°≈à kartu konkr√©tnemu hr√°ƒçovi
                self.hraci[meno_hraca]["karty na ruke"].remove(vyber_kartu)
                self.hraci[meno_hraca]["odhadzovaci balicek"].append(vyber_kartu)
                self.karty_zahrane_v_kole.append((vyber_kartu, meno_hraca)) # potrebne na vyhodnocovanie neskor,a tak aby vedel python precitat ci nebola jedna karta zahrana viac krat a ktorymi hracmi
                #print(f"karty zahrane za kolo: {self.karty_zahrane_v_kole}") # netreba printit, bolo len na testovanie
                #print(f"Karty v odhadzovanom balicku pre {meno}: {info['odhadzovaci balicek']}: {self.odhadzovaci_balicek}")
                #print(f"Karty v odhadzovanom balicku pre {meno_hraca}: {self.hraci[meno_hraca]['odhadzovaci balicek']}") # nepotrebujem uz asi ani
                break # fuu, toto tu treba :D inak sa to nikdy neskonci
    
        
            
        
        
            
    def vyhodnot_kolo(self):
        print("\n=============== VYHODNOTENIE KOLA ===============")
        print("\n=================================================")
        
        # ak si hraci vybrali rovnake karty, vyblokovali a,  tak neziskavaju ziaden bod, poradie sa nemeni
        # potom sa vyhodnotia ostatni hraci, zacne sa s tym, ktory dal kartu s najnizsim cislom, jeho poradie sa posuva pred tych, co sa vyblokovali
        # cize hrac s najvysskou nevyblokovanou kartou by mal byt na prvom mieste
        # hrac co skoncil posledny prichadza o jeden zivot
        # nasledne tento posledny hrac ide na prve miesto, a posuva ostatnych o priecku nizssie   
             
        #najdi vyblokovane karty, najdem si co sa mi opakovalo . tym co sa opakovali sa nateraz nebude menit poradie
                # 1.Vytvor√≠me zoznam v≈°etk√Ωch druh√Ωch hodn√¥t (index 0 v ka≈ædej n-tici)
        cisla = [karta[0] for karta in self.karty_zahrane_v_kole]
                # 2. Pou≈æite Counter na zistenie frekvencie ƒç√≠sel
        pocet_cisel = Counter(cisla)
                # 3. Identifikujte tie ƒç√≠sla, ktor√© sa duplikuj√∫ (maj√∫ poƒçet > 1)
        duplikatne_cisla = {cisla for cisla, pocet in pocet_cisel.items() if pocet > 1}
                # 4. Iterujte cez p√¥vodn√Ω zoznam a vyberte karty, ktor√© maj√∫ duplicitn√© ƒç√≠slo
        karty_s_duplikatnym_cislom = [karta for karta in self.karty_zahrane_v_kole if karta[0] in duplikatne_cisla]
        #s tymto zapisom to nefungovalo spravne: zoradene_duplikatne_karty = sorted(karty_s_duplikatnym_cislom, key=lambda x: x[0], reverse = False)
        zoradene_duplikatne_karty = sorted(
        karty_s_duplikatnym_cislom, 
        key=lambda x: self.hraci[x[1]]["poradie"]  # x[1] je meno hr√°ƒça
    )
        
        neduplikatne_karty = [karta for karta in self.karty_zahrane_v_kole if karta[0] not in duplikatne_cisla]
        #print(f"Neduplik√°tne karty: {neduplikatne_karty}") #testovanie metody, preto print
        
        #print(f"Duplik√°tne ƒç√≠sla s√∫: {duplikatne_cisla}") #testovanie metody, preto print
        print(f"Karty s duplik√°tnymi ƒç√≠slami: {karty_s_duplikatnym_cislom}")
        
         # KROK 3: Zoraƒè neduplik√°tne karty od najvy≈°≈°ej po najni≈æ≈°iu
        zoradene_karty = sorted(neduplikatne_karty, key=lambda x: x[0], reverse=True)
        
       
        
        #prejdi zoradene karty, pre poradie, zaciname od 1.. zoberie si kartu s najvyssim poctom  bodov
        # cize na indexe nula budeme mat prve miesto. aby sme ale realne printili od jednotky a nie nuly
        # vytvorime poradie s hodnotou jedna, spravime for loop
        #pre kazdu srandu zo zoradenych kariet, (karta, meno hraca) , vyprinti info vetu ako sme natom
        # a nezabudni priradit tymto hracom nove poradie 
        
        poradie = 1
        for item in zoradene_karty:
            karta = item[0]
            meno_hraca = item[1]
            print(f"{poradie}. miesto: {meno_hraca} zahral kartu {karta}")               
            self.hraci[meno_hraca]["poradie"] = poradie         
            poradie += 1
        
         
            
            
            
    #aby som mohla najskor priradit poradie duplikatom, tak musim dat pozor na ich povodne poradie
    #vysortovat od mensieho po vacsie cislo podla poradia
    #priradit za poradie neduplikatov
        poradie_duplikatov = poradie
        for item in zoradene_duplikatne_karty:
            karta = item[0]
            meno_hraca = item[1]
            print(f"{poradie_duplikatov}. miesto: {meno_hraca} zahral kartu {karta}")
            self.hraci[meno_hraca]["poradie"] = poradie_duplikatov #musi tu byt takto, aby potom priradilo aj poradia do self.hraci
            poradie_duplikatov += 1 
            
        print("\n üèä‚Äç‚ôÇüèä‚Äç‚ôÇüèä‚Äç‚ôÇüèä‚Äç‚ôÇü¶àüèä‚Äç‚ôÇüèä‚Äç‚ôÇüèä‚Äç‚ôÇüèä‚Äç‚ôÇü¶àüèä‚Äç‚ôÇüèä‚Äç‚ôÇüèä‚Äç‚ôÇüèä‚Äç‚ôÇü¶àüèä‚Äç‚ôÇüèä‚Äç‚ôÇüèä‚Äç‚ôÇüèä‚Äç‚ôÇü¶àüèä‚Äç‚ôÇüèä‚Äç‚ôÇüèä‚Äç‚ôÇüèä‚Äç‚ôÇü¶àüèä‚Äç‚ôÇüèä‚Äç‚ôÇüèä‚Äç‚ôÇüèä‚Äç‚ôÇü¶àüèä‚Äç‚ôÇüèä‚Äç‚ôÇüèä‚Äç‚ôÇüèä‚Äç‚ôÇü¶àüèä‚Äç‚ôÇüèä‚Äç‚ôÇüèä‚Äç‚ôÇüèä‚Äç‚ôÇü¶àüèä‚Äç‚ôÇüèä‚Äç‚ôÇüèä‚Äç‚ôÇüèä‚Äç‚ôÇü¶àüèä‚Äç‚ôÇüèä‚Äç‚ôÇüèä‚Äç‚ôÇüèä‚Äç‚ôÇü¶àüèä‚Äç‚ôÇüèä‚Äç‚ôÇüèä‚Äç‚ôÇüèä‚Äç‚ôÇü¶à")         

                
        
        
        
            
            
                   
    
            
    def uber_zivot(self):
    # zorad hracov podla poradia, od najlepsieho po najhorsieho, x[1] je indexovanie/hodnota kluca -poradie z dictionary
    #meno hraca je nula, poradie 1, musime zoradit podla poradia cize preto x[1]
    #metoda items(), berie dvojice z dictionary
        zoradeni_hraci = sorted(self.hraci.items(), key=lambda x: x[1]["poradie"])
        #print(f" zoradeni hraci: {zoradeni_hraci}")  #testovaci print, v hre to uz len zavadza :D
        
    
    # Posledny hrac (index -1) straca zivot. Treba indexovat kedze neskor pocas hry hracov bude ubudat
    #-1 = posledny prvok v zozname
    #[0] = prv√Ω prvok z dvojice (meno hraca) / zaujima ma ktory HRAC straca zivot
        meno_posledneho = zoradeni_hraci[-1][0]
        self.hraci[meno_posledneho]["zivoty"] -= 1
        print(f"\n ‚ùå‚ùå‚ùå {meno_posledneho} prisiel o zivot! zostava {self.hraci[meno_posledneho]['zivoty']} zivoty/zivot.‚ùå‚ùå‚ùå")
        
        
        
        if self.hraci[meno_posledneho]["zivoty"] == 0:
            return meno_posledneho  # vrat meno vypadnuteho hraca, ak je teda true, tak sa vrati meno a funkcia okmazite konci. nebude oznamovat ze sa posuva na dalsie miesto
            
        
        return None  # Nikto nevypadol
        
    
    #po odratani zivota ide posledny hrac na prve miesto a posuva ostatnych hracov na              
    def posledny_na_prve(self): 
        # Zorad hracov podla aktualneho poradia       
        zoradeni_hraci2 = sorted(self.hraci.items(), key=lambda x: x[1]["poradie"])
        
        # Posledny hrac (ten s najhorsim poradim)
        meno_posledneho2 = zoradeni_hraci2[-1][0]
        # Posledny ide na prve miesto (poradie = 1), na tvrdo zadefinovane
        self.hraci[meno_posledneho2]["poradie"] = 1
        
        for item in range(len(zoradeni_hraci2) - 1): #posledneho mame uz supnuteho na prve miesto, cize nam zostava zvysok hracov minus jeden
            meno_hraca = zoradeni_hraci2[item][0]
            self.hraci[meno_hraca]["poradie"] += 1
            
    
            
        print(f"\n {meno_posledneho2} ide na prve miesto, ostatni sa posuvaju.")
        
        
        
        
        
    ############################################################################################    
        
    
    def dopln_karty(self):
        # pozri hracov, ak ma niekto na ruke menej ako 3 karty, 
        # zober karty z odhadozovacieho balika, a dopln ich do ruky
        
        for meno, info in self.hraci.items():
            karty_na_ruke = info["karty na ruke"]
            odhadzovaci_balik = info["odhadzovaci balicek"]
            
            if len(karty_na_ruke) < 3:
                print(f"{meno} ma len {len(karty_na_ruke)} karty, doplna si z odhadzovaciho balika")
                karty_na_ruke.extend(odhadzovaci_balik) #prida z odhadzovacieho balika do kariet na ruke
                odhadzovaci_balik.clear()  # Vyprazdni odhadzovaci balik
            
                #print(f"{meno} dobera karty a ma plny pocet kariet") #netreba mi?
                
                
          
                
                 
   ###################################################################################################     
    def aktivni_hraci(self):
     #zoznam mien hracov, co maju aktivne zivoty
        hrajuci_hraci = []  # Zoznam, nie dictionary... mena su dolezite, staci zonam, na tie mena sa potom viazu ostatne vlastnosti
        vypadnuti = []
        for meno, info in self.hraci.items():  # items() da dvojice, 
            pocet_zivotov = info["zivoty"]  # Takto ceknem zivoty
        
            if pocet_zivotov > 0:
                hrajuci_hraci.append(meno)  # Pridaj meno do zoznamu
            else:                
                vypadnuti.append(meno)  # Poznaƒç si koho odstr√°ni≈°
            
        #vyhod vypadnutych von z dictionary
        for meno in vypadnuti:
            del self.hraci[meno] # metoda clear by mi vymazala vsetkych, pouzijeme prikaz vymaz meno z vypadnutych zo slovniku
            print(f"{meno} vypadol z hry!")       
                
                
    
        #print(f"Hrajuci hraci: {hrajuci_hraci}") #toto tiez vykomentovavam, bude print v play, uvidime co to spravi
       
        return hrajuci_hraci  # return mimo for loopu. ak by to bolo v ramci for loopu, nechecklo by to vsetkych hracov, hned by sa to zastavilo
                              # tym ze je tu return, viem dostat tuto premennu aj von z funkcie. ak by nebol return, funkcia skonci a premenna zanikne
    
    def hraj_kolo(self):
        print("\n=============== NOVE KOLO ===============")
    ##### PRIEBEH HRY #######
    
    
        self.karty_zahrane_v_kole = []
    # 1. zisti ktori hraci su aktivni, zavolaj funkciu ativini_hraci a uloz adta do premennej aktivni
        aktivni = self.aktivni_hraci() #treba rozbalit oboch, kedze  sme v metode returnovali obe hodnoty
       
        print(f"V tomto kole hraju: {', '.join(aktivni)}") #.join mi pekne vyprinti hracov bez zatvoriek
        
    
    # 2. Zoraƒè ich podƒæa poradia (pou≈æij sorted)
    
        print("Poradie na zaciatku kola")
        hraci_podla_poradia = sorted(self.hraci.items(), key=lambda x: x[1]["poradie"])
        for meno, info in hraci_podla_poradia:
            print(f"{info['poradie']}. miesto: {meno}")
        
        
    
    # 3. kazdy aktivny hrac zahra kartu (for loop + zahraj_kartu)
        for hraci in aktivni:
            self.zahraj_kartu(hraci)
            if not self.hra_bezi:  # Ak niekto ukoncil hru
                return  # skonci kolo ak niekto ukoncil hru
        
                
    
    # 4. Vyhodno≈• kolo
        self.vyhodnot_kolo()
    
    # 5. Uber ≈æivot
        vypadnuti_hraci = self.uber_zivot()
        if vypadnuti_hraci is None:
            self.posledny_na_prve()  # Posun sa vykona len ak hrac este zije
        else:
            print(f"ü¶à {vypadnuti_hraci} vypadol z hry!")
            del self.hraci[vypadnuti_hraci]  # Vyma≈æ vypadnut√©ho hr√°ƒça
    # ≈ΩIADNE pos√∫vanie - poradie zost√°va ako je


    
    # 7. dopln karty hracom co maju dve a menej kariet
        self.dopln_karty()
        
        
    
    def play(self):
        print("Vitaj v hre Chramst!")
        print("pravidla su jednoduche: nenechaj sa zozrat zralokom")
        print("kazde kolo hrac zahra kartu")
        print("karta s najvyssim cislom posuva hraca na prve miesto")
        print("ale pozor, ak hraci zahraju rovnake karty, tak sa z povodneho miesta nehybu a moze ich predbehnut hrac s nizsou kartou")
        print("na konci kola, hrac na poslednom mieste straca zivot - zralok mu odkusne jednu z koncatin,  a posuva sa na prve miesto, zacina sa dalsie kolo")
        print("hra konci ak zostanu dvaja hraci, vyhrava ten co je na prvom mieste")
        
        while len(self.aktivni_hraci()) > 2 and self.hra_bezi:  # hraj pokial mame viac ako dvoch hracov
            self.hraj_kolo()
            
        if self.hra_bezi:  # len ak sa hra skoncila normalne
            zoradeni = sorted(self.hraci.items(), key=lambda x: x[1]["poradie"])
            vitaz = zoradeni[0][0]  # prvy v zoradenom zozname je vitaz
            print(f"\nüéâüéâüéâ HRA SKONCILA! VITAZ JE: {vitaz}! üéâüéâüéâ")
        else:
            print("\nüëã Hra bola ukoncena hracom.")
    
        
               

    



            
       
        
        
        
        
        
        


# Vytvorenie instance hry a zahranie karty
hra = Chramst()
hra.play()






```
