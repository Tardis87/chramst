```python
from collections import Counter


class Chramst:
    def __init__(self):
        
        self.hraci = {
            "Modry": {"poradie": 1, "zivoty": 4, "karty na ruke": [1, 2, 3, 4, 5, 6], "odhadzovaci balicek": []},
            "Zlty": {"poradie": 2, "zivoty": 4, "karty na ruke": [1, 2, 3, 4, 5, 6], "odhadzovaci balicek": []},
            "Zeleny": {"poradie": 3, "zivoty": 4, "karty na ruke": [1, 2, 3, 4, 5, 6], "odhadzovaci balicek": []},
            "Cerveny": {"poradie": 4, "zivoty": 4, "karty na ruke": [1, 2, 3, 4, 5, 6], "odhadzovaci balicek": []}
        }
        
        self.karty_zahrane_v_kole = []
        
    def zobraz_hracov(self):
        for meno, info in self.hraci.items():
            print(f"{meno}: {info['poradie']}. miesto, Zivoty: {info['zivoty']}. karty na ruke: {info['karty na ruke']}. karty v odhadzovacom balicku: {info['odhadzovaci balicek']} ")
         
     
        
    def zahraj_kartu(self, meno_hraca):
        # Zisti karty tohto hraca
        karty_hraca = self.hraci[meno_hraca]["karty na ruke"]
        vyber_kartu = int(input(f"{meno_hraca}, vyber si kartu co chces zahrat {karty_hraca}: "))
       
        
        
        
        if vyber_kartu not in karty_hraca:
            print(f"Cislo {vyber_kartu} nie je spravne. Vyber si jednu z kariet co mas na ruke")
        else:
            print(f"{meno_hraca} zahral kartu {vyber_kartu}")
            # Odstráň kartu konkrétnemu hráčovi
            self.hraci[meno_hraca]["karty na ruke"].remove(vyber_kartu)
            self.hraci[meno_hraca]["odhadzovaci balicek"].append(vyber_kartu)
            self.karty_zahrane_v_kole.append((vyber_kartu, meno_hraca)) # potrebne na vyhodnocovanie neskor,a tak aby vedel python precitat ci nebola jedna karta zahrana viac krat a ktorymi hracmi
            print(f"karty zahrane za kolo: {self.karty_zahrane_v_kole}")
            #print(f"Karty v odhadzovanom balicku pre {meno}: {info['odhadzovaci balicek']}: {self.odhadzovaci_balicek}")
            print(f"Karty v odhadzovanom balicku pre {meno_hraca}: {self.hraci[meno_hraca]['odhadzovaci balicek']}")
            
    
        
            
        
        
            
    def vyhodnot_kolo(self):
        print("\n=== VYHODNOTENIE KOLA ===")
        
        # ak si hraci vybrali rovnake karty, vyblokovali a,  tak neziskavaju ziaden bod, poradie sa nemeni
        # potom sa vyhodnotia ostatni hraci, zacne sa s tym, ktory dal kartu s najnizsim cislom, jeho poradie sa posuva pred tych, co sa vyblokovali
        # cize hrac s najvysskou nevyblokovanou kartou by mal byt na prvom mieste
        # hrac co skoncil posledny prichadza o jeden zivot
        # nasledne tento posledny hrac ide na prve miesto, a posuva ostatnych o priecku nizssie   
             
        #najdi vyblokovane karty, najdem si co sa mi opakovalo . tym co sa opakovali sa nateraz nebude menit poradie
                # 1.Vytvoríme zoznam všetkých druhých hodnôt (index 0 v každej n-tici)
        cisla = [karta[0] for karta in self.karty_zahrane_v_kole]
                # 2. Použite Counter na zistenie frekvencie čísel
        pocet_cisel = Counter(cisla)
                # 3. Identifikujte tie čísla, ktoré sa duplikujú (majú počet > 1)
        duplikatne_cisla = {cisla for cisla, pocet in pocet_cisel.items() if pocet > 1}
                # 4. Iterujte cez pôvodný zoznam a vyberte karty, ktoré majú duplicitné číslo
        karty_s_duplikatnym_cislom = [karta for karta in self.karty_zahrane_v_kole if karta[0] in duplikatne_cisla]
        zoradene_duplikatne_karty = sorted(karty_s_duplikatnym_cislom, key=lambda x: x[0], reverse = False)
        
        neduplikatne_karty = [karta for karta in self.karty_zahrane_v_kole if karta[0] not in duplikatne_cisla]
        print(f"Neduplikátne karty: {neduplikatne_karty}")
        
        print(f"Duplikátne čísla sú: {duplikatne_cisla}")
        print(f"Karty s duplikátnymi číslami: {karty_s_duplikatnym_cislom}")
        
         # KROK 3: Zoraď neduplikátne karty od najvyššej po najnižšiu
        zoradene_karty = sorted(neduplikatne_karty, key=lambda x: x[0], reverse=True)
        
       
        
        #prejdi zoradene karty, pre poradie, zaciname od 1.. zoberie si kartu s najvyssim poctom  bodov
        # cize na indexe nula budeme mat prve miesto. aby sme ale realne printili od jednotky a nie nuly
        # vytvorime poradie s hodnotou jedna, spravime for loop
        #pre kazdu srandu zo zoradenych kariet, (karta, meno hraca) , vyprinti info vetu ako sme natom
        # a nezabudni priradit tymto hracom nove poradie 
        
        poradie = 1
        for item in zoradene_karty:
            karta = item[0]
            meno_hraca = item[1]
            print(f"{poradie}. miesto: {meno_hraca} zahral kartu {karta}")               
            self.hraci[meno_hraca]["poradie"] = poradie         
            poradie += 1
            
            
            
            
    #aby som mohla najskor priradit poradie duplikatom, tak musim dat pozor na ich povodne poradie
    #vysortovat od mensieho po vacsie cislo podla poradia
    #priradit za poradie neduplikatov
        poradie_duplikatov = poradie
        for item in zoradene_duplikatne_karty:
            karta = item[0]
            meno_hraca = item[1]
            print(f"{poradie_duplikatov}. miesto: {meno_hraca} zahral kartu {karta}")
            self.hraci[meno_hraca]["poradie"] = poradie_duplikatov #musi tu byt takto, aby potom priradilo aj poradia do self.hraci
            poradie_duplikatov += 1 
            
                 
                
    
            
    def uber_zivot(self):
    # zorad hracov podla poradia, od najlepsieho po najhorsieho, x[1] je indexovanie/hodnota kluca -poradie z dictionary
    #meno hraca je nula, poradie 1, musime zoradit podla poradia cize preto x[1]
    #metoda items(), berie dvojice z dictionary
        zoradeni_hraci = sorted(self.hraci.items(), key=lambda x: x[1]["poradie"])
        print(f" zoradeni hraci: {zoradeni_hraci}")
    
    # Posledny hrac (index -1) straca zivot. Treba indexovat kedze neskor pocas hry hracov bude ubudat
    #-1 = posledny prvok v zozname
    #[0] = prvý prvok z dvojice (meno hraca) / zaujima ma ktory HRAC straca zivot
        meno_posledneho = zoradeni_hraci[-1][0]
        self.hraci[meno_posledneho]["zivoty"] -= 1
        print(f"{meno_posledneho} prisiel o zivot! zostava {self.hraci[meno_posledneho]['zivoty']} zivoty/zivot.")
        
        
    
    #po odratani zivota ide posledny hrac na prve miesto a posuva ostatnych hracov na              
    def posledny_na_prve(self): 
        # Zorad hracov podla aktualneho poradia       
        zoradeni_hraci2 = sorted(self.hraci.items(), key=lambda x: x[1]["poradie"])
        
        # Posledny hrac (ten s najhorsim poradim)
        meno_posledneho2 = zoradeni_hraci2[-1][0]
        # Posledny ide na prve miesto (poradie = 1), na tvrdo zadefinovane
        self.hraci[meno_posledneho2]["poradie"] = 1
        
        for item in range(len(zoradeni_hraci2) - 1): #posledneho mame uz supnuteho na prve miesto, cize nam zostava zvysok hracov minus jeden
            meno_hraca = zoradeni_hraci2[item][0]
            self.hraci[meno_hraca]["poradie"] += 1
    
            
        print(f"{meno_posledneho2} ide na prve miesto, ostatni sa posuvaju.")
        
        
    def aktivni_hraci(self):
    # zoznam mien hracov, co maju aktivne zivoty
        hrajuci_hraci = []  # Zoznam, nie dictionary... mena su dolezite, staci zonam, na tie mena sa potom viazu ostatne vlastnosti
    
        for meno, info in self.hraci.items():  # items() da dvojice, 
            pocet_zivotov = info["zivoty"]  # Takto ceknem zivoty
        
            if pocet_zivotov > 0:
                hrajuci_hraci.append(meno)  # Pridaj meno do zoznamu
    
        print(f"Hrajuci hraci: {hrajuci_hraci}")
        return hrajuci_hraci  # return mimo for loopu. ak by to bolo v ramci for loopu, nechecklo by to vsetkych hracov, hned by sa to zastavilo
                              # tym ze je tu return, viem dostat tuto premennu aj von z funkcie. ak by nebol return, funkcia skonci a premenna zanikne
    
    def hraj_kolo(self):
        print("\n=== NOVE KOLO ===")
    ##### PRIEBEH HRY #######
    # 1. zisti ktori hraci su aktivni, zavolaj funkciu ativini_hraci a uloz adta do premennej aktivni
        aktivni = self.aktivni_hraci()
        print(f"V tomto kole hraju(hraj_kolo): {aktivni}")
    
    # 2. Zoraď ich podľa poradia (použij sorted)
        hraci_podla_poradia = sorted(self.hraci.items(), key=lambda x: x[1]["poradie"])
        print(f"poradie hracov: {hraci_podla_poradia}")
        
        
    
    # 3. kazdy aktivny hrac zahra kartu (for loop + zahraj_kartu)
        for hraci in aktivni:
            self.zahraj_kartu(hraci)
                
    
    # 4. Vyhodnoť kolo
        self.vyhodnot_kolo()
    
    # 5. Uber život
        self.uber_zivot()
    
    # 6. Posledný na prvé
        self.posledny_na_prve()
    
    # 7. Vyčisti karty_zahrane_v_kole = [] (priprav sa na nové kolo)
        
    
    def play(self):
        print("Vitaj v hre Chramst!")
        
        while len(self.aktivni_hraci()) > 2:  # Pokým je viac ako 12 hráč
            self.hraj_kolo()
    
    
               

    
# Zivoty idu do minus jedna, treba nastavit aby to removlo hraca co ma nula zivotov
# poradie sa nerestejue, kazde kolo poradie v range poctu aktivnych hracov
# karty z odhadzovacieho balicku: ak ostane na ruke iba jedna karta , hrac si dolozi vsetky ostatne karty


            
       
        
        
        
        
        
        


# Vytvorenie instance hry a zahranie karty
hra = Chramst()
hra.play()






```
